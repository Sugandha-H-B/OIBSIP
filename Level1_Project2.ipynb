```python
df1['AvgPurchase']=df1['MntTotal']/df1['AcceptedCmpOverall'].replace(0,1)
```

```python
df1['PurchaseFrequency']=df1['AcceptedCmpOverall']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
features=df1[['Income','Recency','MntTotal','AvgPurchase','PurchaseFrequency']]
```

```python
scaler=StandardScaler()
scaled_features=scaler.fit_transform(features)
```

```python
inertia=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,random_state=42)
    df1['Data_obt']=kmeans.fit_predict(scaled_features)
    inertia.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
# Apply KMeans with optimal k (say, k=4)
kmeans = KMeans(n_clusters=4, random_state=42)
df1['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df1, x='Income', y='MntTotal', hue='Cluster', palette='viridis')
plt.title('Customer Segments by Income and Spending')
plt.show()
```

```python
# Bar chart of average spending per cluster
df1.groupby('Cluster')['MntTotal'].mean().plot(kind='bar', title='Avg Spending per Cluster')
plt.ylabel('Average Spending')
plt.show()
```
