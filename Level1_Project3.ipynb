```python
numerical_columns=['price','minimum_nights','number_of_reviews','calculated_host_listings_count','availability_365']
stats=data[numerical_columns].describe()
stats
```

```python
data[numerical_columns].median()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
for col in numerical_columns:
    Q1=data[col].quantile(0.25)
    Q3=data[col].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    outliers=data[(data[col]<lower_bound)|(data[col]>upper_bound)]
    sns.boxplot(data[col],color='red')
    plt.title(f'Boxplot for {col}with outliers')
    plt.show()
    
```

```python
import plotly.express as px
from sklearn.tree import DecisionTreeRegressor,plot_tree
from sklearn.model_selection import train_test_split
data['estimated_revenue']=data['price']*data['availability_365']
```

```python
corr=data[['price','minimum_nights','number_of_reviews','calculated_host_listings_count','availability_365','estimated_revenue']].corr()
sns.heatmap(corr,annot=True,cmap='viridis')
plt.title("Correlation Matrix")
```

```python
X=data[numerical_columns]
y=data['estimated_revenue']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
tree=DecisionTreeRegressor(max_depth=4,random_state=0)
tree.fit(X_train,y_train)
plt.figure(figsize=(30,10))
plot_tree(tree,feature_names=X.columns,filled=True)
plt.show()
```

```python
importances=pd.Series(tree.feature_importances_,index=X.columns)
importances.sort_values().plot(kind='barh')
plt.title("Feature Importance")
```
